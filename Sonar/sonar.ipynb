```python
#Import Packages
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Read the data file
df = pd.read_csv('sonar_data.csv', header=None)
```

```python
#Show the first few rows of data
df.head()
```

```python
#Tell the shape of data
df.shape
```

```python
# Describe the statistical quantities for each column
df.describe()
```

```python
#Find the number of rows of each type 
df[60].value_counts().plot(kind = 'pie')
```

```python
correlation = df.corr()
sns.heatmap(correlation, cmap = 'Greens')
```

```python
# Find the mean of parameters of each type
df.groupby(60).mean()
```

```python
# Set the independent and dependent variable
x = df.drop(columns=60, axis =1)
y = df[60]
print(x, y)
```

```python
# Training and Testing data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.1, stratify = y, random_state = 1)
```

```python
#Show the trained data 
print(xtrain, ytrain)
```

```python
#Show the test data 
print(xtest, ytest)
```

```python
# Set the model to be used
model = LogisticRegression()
```

```python
#Training the model
model.fit(xtrain, ytrain)
```

```python
#Model Evaluation
#Training accuracy
xtrain_prediction = model.predict(xtrain)
training_accuracy = accuracy_score(xtrain_prediction, ytrain)
print(training_accuracy)
#Test accuracy
xtest_prediction = model.predict(xtest)
test_accuracy = accuracy_score(xtest_prediction, ytest)
print(test_accuracy)
```
